```python
# 2 by 2 subplot 
fig, axes= plt.subplots(2,2, figsize=(25,11))
#create scatter plot for Revenue vs visitors 
bar_plot=sns.barplot(x='Species', y='Sample Number', data=df, ci=False, ax=axes[0,0])
bar_plot.set_title('Bar plot for Species and Sample Number')

#bar plot for day_name vs Revenue
#sns.barplot(x='Day_Name', y='Revenue', data=raw_data, ax=axes[0,1])
# Calculating the mean
mean=df['Body Mass (g)'].mean()
#ploting the mean
#plt.axvline(mean, color='red',ax=axes[0,1])
hist_plot=df['Body Mass (g)'].values
histtt=sns.histplot(hist_plot,ax=axes[0,1])
histtt.set_title('Histogram for Body Mass')


#histgram for Revenue

box_plot=sns.boxplot(x='Species', y=df['Body Mass (g)'].values,
                     data=df, ax=axes[1,0])
box_plot.set_title('Box Plot for Species and Body Mass')
box_plot.set_ylabel('Body Mass')
#line plot for Week_ID vs revenue

scatter_plot=sns.scatterplot(x='Culmen Depth (mm)', y='Body Mass (g)',
                     data=df, color='red', hue='Species',ax=axes[1,1])
scatter_plot.set_title('Scatter Plot for Culmen Depth and Body Mass')
plt.savefig('snsplot.png')
```
